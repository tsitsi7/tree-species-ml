```python
!pip install -q icrawler

from icrawler.builtin import BingImageCrawler
import os

DOWNLOAD_ROOT = "/content/drive/MyDrive/content/fruit_trees" 

queries = {
    "Banana":  "banana tree top view aerial canopy",
    "Mango":   "mango tree top view aerial canopy",
    "Coconut": "coconut palm tree top view aerial",
}

for species, query in queries.items():
    save_dir = os.path.join(DOWNLOAD_ROOT, species)
    os.makedirs(save_dir, exist_ok=True)
    crawler = BingImageCrawler(storage={"root_dir": save_dir})
    crawler.crawl(keyword=query, max_num=120)   # grab 120, keep best 100
    print(f"{species}: {len(os.listdir(save_dir))} images downloaded")
```
